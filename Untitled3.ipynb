```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture
import joblib
from google.colab import drive

drive.mount('/content/drive')

training_images = [
    "/content/drive/My Drive/images/water_body_1.jpg",
    "/content/drive/My Drive/images/water_body_2.jpg",
    "/content/drive/My Drive/images/water_body_27.jpg",
    "/content/drive/My Drive/images/water_body_39.jpg",
    "/content/drive/My Drive/images/water_body_78.jpg",
    "/content/drive/My Drive/images/water_body_120.jpg",
    "/content/drive/My Drive/images/water_body_21.jpg",
    "/content/drive/My Drive/images/water_body_100.jpg",
    "/content/drive/My Drive/images/water_body_301.jpg",
    "/content/drive/My Drive/images/water_body_781.jpg",
    "/content/drive/My Drive/images/water_body_170.jpg",
    "/content/drive/My Drive/images/water_body_22.jpg",
    "/content/drive/My Drive/images/water_body_273.jpg",
    "/content/drive/My Drive/images/water_body_391.jpg",
    "/content/drive/My Drive/images/water_body_782.jpg",
    "/content/drive/My Drive/images/water_body_1204.jpg",
    "/content/drive/My Drive/images/water_body_216.jpg",
    "/content/drive/My Drive/images/water_body_276.jpg",
    "/content/drive/My Drive/images/water_body_398.jpg",
    "/content/drive/My Drive/images/water_body_781.jpg",
]

test_images = [
    "/content/drive/My Drive/images/water_body_21.jpg",
    "/content/drive/My Drive/images/water_body_27.jpg",
    "/content/drive/My Drive/images/water_body_39.jpg",
    "/content/drive/My Drive/images/water_body_78.jpg",
]

model_path = "/content/drive/My Drive/gmm_water_model.pkl"

def prepare_training_data(image_paths):
    """ Load multiple images, convert to Lab space, and create a dataset of pixels. """
    pixel_data = []

    for img_path in image_paths:
        image = cv2.imread(img_path)
        if image is None:
            print(f"Error loading {img_path}, skipping...")
            continue

        image_lab = cv2.cvtColor(image, cv2.COLOR_BGR2LAB)
        pixels = image_lab.reshape(-1, 3)
        pixel_data.append(pixels)

    return np.vstack(pixel_data)

def train_gmm(pixel_data, n_components=3):
    """ Train a GMM model to classify water and non-water pixels. """
    gmm = GaussianMixture(n_components=n_components, random_state=42)
    gmm.fit(pixel_data)
    return gmm

def detect_water(image, gmm):
    """ Predict water areas in an image using a trained GMM model. """
    img_lab = cv2.cvtColor(image, cv2.COLOR_BGR2LAB)
    img_reshaped = img_lab.reshape((-1, 3))

    labels = gmm.predict(img_reshaped)

    water_cluster = np.argmin(gmm.means_[:, 1])
    water_mask = (labels == water_cluster).astype(np.uint8) * 255

    return water_mask.reshape(image.shape[:2])

try:
    print("Loading trained model...")
    gmm = joblib.load(model_path)
    print("Model loaded successfully!")
except:
    print("Training new model...")
    training_data = prepare_training_data(training_images)
    gmm = train_gmm(training_data)
    joblib.dump(gmm, model_path)
    print("Model trained and saved!")

for image_path in test_images:
    image = cv2.imread(image_path)
    if image is None:
        print(f"Error loading {image_path}, skipping...")
        continue

    image = cv2.resize(image, (512, 512))
    water_mask = detect_water(image, gmm)

    plt.figure(figsize=(12, 6))

    plt.subplot(1, 2, 1)
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title("Original Image")

    plt.subplot(1, 2, 2)
    plt.imshow(water_mask, cmap="gray")
    plt.title("Water Detection (GMM)")

    plt.show()
```
