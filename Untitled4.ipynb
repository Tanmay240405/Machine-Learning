```python
import cv2
import numpy as np
import skimage.segmentation as seg
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture
from google.colab import drive


drive.mount('/content/drive')

def superpixel_segmentation(image):
    """ Perform superpixel segmentation using SLIC. """
    img_hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)


    lower_blue = np.array([90, 50, 50])
    upper_blue = np.array([140, 255, 255])
    blue_mask = cv2.inRange(img_hsv, lower_blue, upper_blue)


    segments = seg.slic(image, n_segments=300, compactness=10, sigma=1, start_label=1)

    return segments, blue_mask

def detect_water(image, n_components=3):
    """ Identify water regions using GMM in LAB color space. """
    img_lab = cv2.cvtColor(image, cv2.COLOR_BGR2LAB)
    img_reshaped = img_lab.reshape((-1, 3))

    gmm = GaussianMixture(n_components=n_components, random_state=42)
    gmm.fit(img_reshaped)
    labels = gmm.predict(img_reshaped)

    segmented = labels.reshape(image.shape[:2])


    water_mask = (segmented == np.argmin(gmm.means_[:, 0])).astype(np.uint8) * 255

    return water_mask

def process_image(image_path):
    """ Full pipeline: segmentation and water extraction. """
    image = cv2.imread(image_path)
    if image is None:
        raise ValueError(f"Error loading image: {image_path}")

    image = cv2.resize(image, (512, 512))

    segments, blue_mask = superpixel_segmentation(image)

    water_mask = detect_water(image)

    plt.figure(figsize=(12, 8))

    plt.subplot(1, 2, 1)
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title("Original Image")

    plt.subplot(1, 2, 2)
    plt.imshow(water_mask, cmap="gray")
    plt.title("Detected Water Bodies")

    plt.show()

image_files = [
    '/content/drive/My Drive/images/water_body_120.jpg',
    '/content/drive/My Drive/images/water_body_21.jpg',
    '/content/drive/My Drive/images/water_body_27.jpg',
    '/content/drive/My Drive/images/water_body_39.jpg'
]

for image_path in image_files:
    process_image(image_path)
```
